```python
import pymongo

# Connect to the MongoDB server running on localhost
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create or switch to the 'example' database
db = client["example"]

# Create or switch to the 'books' collection
collection = db["books"]

# Sample data
books_data = [
    {"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "year": 1925},
    {"title": "To Kill a Mockingbird", "author": "Harper Lee", "year": 1960},
    {"title": "1984", "author": "George Orwell", "year": 1949},
    {"title": "The Catcher in the Rye", "author": "J.D. Salinger", "year": 1951},
]

# Insert the sample data into the 'books' collection
result = collection.insert_many(books_data)

# Print the IDs of the inserted documents
print("Inserted IDs:", result.inserted_ids)

# Close the MongoDB connection
client.close()
```
